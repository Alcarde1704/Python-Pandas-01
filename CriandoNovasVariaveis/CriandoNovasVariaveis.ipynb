## Calculando valor bruto e adicionando ao dataframe

```python
dados['Valor Bruto'] = (dados['Valor'] * 12) + (dados['Condominio'] * 12)
```

```python
dados['Valor Bruto Mensal'] = dados['Valor'] + dados['Condominio']
```

```python
dados
```

## Calculando valor do metro quadrado

```python
dados['Valor m2'] = dados['Valor'] / dados['Area']
```

```python
dados
```

```python
dados['Valor m2'] = dados['Valor m2'].round(2) # arredondar em 2 casas
```

```python
dados
```

```python
dados['Valor Bruto Mensal m2'] = (dados['Valor Bruto Mensal'] / dados['Area']).round(2)
```

```python
dados
```

```python
casa = ['Casa', 'Casa de Condominio', 'Casa de Vila']
```

```python
dados['Tipo agregado'] = dados['Tipo'].apply(lambda x: 'Casa' if x in casa else 'Apartamento')
```
