## Estatisticas Descritivas

```python
grupo_bairros['Valor'].describe().round(2)
```

```python
grupo_bairros['Valor'].aggregate(['min', 'max', 'sum']).rename(columns = {'min': 'Minimo', 'max': 'Máximo', 'sum': 'Soma'})
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.rc('figure', figsize = (20,10))
```

```python
fig = grupo_bairros['Valor'].mean().plot.bar(color = 'blue')
fig.set_ylabel('Valor Do Aluguel')
fig.set_title('Valor Médio do Aluguel por Bairro', {'fontsize': 22})
```

```python
fig = grupo_bairros['Valor'].mean().plot.bar(color = 'blue')
fig.set_ylabel('Valor Do Aluguel')
fig.set_title('Valor Médio do Aluguel por Bairro', {'fontsize': 22})
```

```python
fig = grupo_bairros['Valor'].max().plot.bar(color = 'blue')
fig.set_ylabel('Valor Do Aluguel')
fig.set_title('Valor Médio do Aluguel por Bairro', {'fontsize': 22})
```

### Atividades 

```python
precos = pd.DataFrame([['Feira', 'Cebola', 2.5], 
                        ['Mercado', 'Cebola', 1.99], 
                        ['Supermercado', 'Cebola', 1.69], 
                        ['Feira', 'Tomate', 4], 
                        ['Mercado', 'Tomate', 3.29], 
                        ['Supermercado', 'Tomate', 2.99], 
                        ['Feira', 'Batata', 4.2], 
                        ['Mercado', 'Batata', 3.99], 
                        ['Supermercado', 'Batata', 3.69]], 
                        columns = ['Local', 'Produto', 'Preço'])
```

```python
precos
```

```python
produtos = precos.groupby('Produto', sort=False)
```

```python
produtos
```

```python
produtos.describe().round(2)
```

```python
produtos['Preço'].aggregate(['mean', 'std', 'min', 'max']).rename(columns = {'mean': 'Média', 'std': 'Desvio Padrão', 'min': 'Mínimo', 'max': 'Máximo'}).round(2)
```

```python
fig1 = produtos['Preço'].aggregate(['mean', 'std', 'min', 'max']).rename(columns = {'mean': 'Média', 'std': 'Desvio Padrão', 'min': 'Mínimo', 'max': 'Máximo'}).round(2).plot.bar()
fig1.set_ylabel('Valores', {'fontsize': 22})
fig1.set_title('Estatisticas Descritivas', {'fontsize': 22})
```
